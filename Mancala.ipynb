```python
import random # randomモジュールを追加

Mancala_P1_holes = [4, 4, 4, 4, 4, 4]
Mancala_P2_holes = [4, 4, 4, 4, 4, 4] #穴のガラス玉の数
Mancala_score_P1 = 0
Mancala_score_P2 = 0 #横のスコアの穴のガラス玉の数

def show_board():
    print("     " + "「 P2: [ 5 ]  [ 4 ]  [ 3 ]  [ 2 ]  [ 1 ]  [ 0 ] 」")
    print("┌─────┬──────────────────────┬───┐")
    print(f"│  (P2) │   {Mancala_P2_holes[5]}      {Mancala_P2_holes[4]}      {Mancala_P2_holes[3]}      {Mancala_P2_holes[2]}      {Mancala_P2_holes[1]}      {Mancala_P2_holes[0]}    │  (P1) │")
    print("│       │                                           │       │")
    print(f"│   {Mancala_score_P2}   │                                           │   {Mancala_score_P1}   │")
    print("│       │                                           │       │")
    print(f"│       │   {Mancala_P1_holes[0]}      {Mancala_P1_holes[1]}      {Mancala_P1_holes[2]}      {Mancala_P1_holes[3]}      {Mancala_P1_holes[4]}      {Mancala_P1_holes[5]}    │       │")
    print("└─────┴──────────────────────┴───┘")
    print("     " + "「 P1: [ 0 ]  [ 1 ]  [ 2 ]  [ 3 ]  [ 4 ]  [ 5 ] 」")

def move(player, player_choice):
    global Mancala_P1_holes, Mancala_P2_holes, Mancala_score_P1, Mancala_score_P2 #グローバル数にしてリストの外から変更できるようにする
    extra_turn = False # 追加ターンを初期化

    if player == "P1":
        if not (0 <= player_choice < len(Mancala_P1_holes)):
            print("無効な穴の番号です。0から5の間で選択してください。") #P1が選択できる穴の判定
            return False, False # 無効な手なのでextra_turnもFalse
        if Mancala_P1_holes[player_choice] == 0:
            print("その穴には石がありません。別の穴を選択してください。") #P1が何も入っていない穴を選んだ判定
            return False, False # 無効な手なのでextra_turnもFalse

        stones_in_hand = Mancala_P1_holes[player_choice] #ガラス玉を手にもつ
        Mancala_P1_holes[player_choice] = 0  #手に取った穴のガラス玉をなくす

        carry_stone_put = player_choice #どこから配り始めるかの基準点を与える

        while stones_in_hand > 0: #手の中のガラス玉がなくなるまで
            carry_stone_put = (carry_stone_put + 1) % 14 #ガラス玉を置く場所をループさせる

            #ここでは、P1の穴を0から5、P1のスコアの穴を6、P2の穴を7から12、P2のスコアの穴を13と表記する

            if carry_stone_put == 13: # P2のスコアの穴はスキップする
                continue

            if 0 <= carry_stone_put <= 5: #P1の穴
                Mancala_P1_holes[carry_stone_put] += 1
            elif carry_stone_put == 6: #P1のスコアの穴
                Mancala_score_P1 += 1
            elif 7 <= carry_stone_put <= 12: #P2の穴
                p2_place = carry_stone_put - 7 #リストに合わせるように引いて合わせる
                Mancala_P2_holes[p2_place] += 1

            stones_in_hand -= 1 #手のガラス玉の中からガラス玉を減少させる

        # P1のキャプチャルールの実装
        # 最後の石が自分の空の穴に落ち、かつ相手の対応する穴に石がある場合
        if stones_in_hand == 0 and 0 <= carry_stone_put <= 5 and Mancala_P1_holes[carry_stone_put] == 1: # 最後に置いた石が、P1の穴で、かつ元々空だった場合（今1つ入っている）
            opposite_hole_index_P2 = 5 - carry_stone_put # P2側の対応する穴のインデックス
            if Mancala_P2_holes[opposite_hole_index_P2] > 0: # 相手の穴に石がある場合
                # 石を捕獲し、P1のスコアに入れる
                Mancala_score_P1 += Mancala_P2_holes[opposite_hole_index_P2] + 1 # 相手の穴の石 + 最後に置いた石
                Mancala_P2_holes[opposite_hole_index_P2] = 0 # 相手の穴を空にする
                Mancala_P1_holes[carry_stone_put] = 0 # 自分が最後に置いた穴を空にする

        if carry_stone_put == 6: # 最後の石がP1の得点ゾーンに入った場合
            extra_turn = True
        return True, extra_turn

    elif player == "P2":
        # P2は自分の穴 (Mancala_P2_holes) から選択する
        if not (0 <= player_choice < len(Mancala_P2_holes)): # P2の穴の範囲をチェック
            return False, False # 失敗を伝えるためにFalseを返す
        if Mancala_P2_holes[player_choice] == 0:
            return False, False # 失敗を伝えるためにFalseを返す

        stones_in_hand = Mancala_P2_holes[player_choice] # P2の穴から石を取る
        Mancala_P2_holes[player_choice] = 0  # 取った穴を空にする

        # P2の選択した穴に対応する全体のボード上の開始位置は 7 + player_choice
        # (P1の穴0-5, P1のスコア6, P2の穴7-12, P2のスコア13) と表記する
        carry_stone_put = 7 + player_choice

        while stones_in_hand > 0: #手の中のガラス玉がなくなるまで
            carry_stone_put = (carry_stone_put + 1) % 14 #ガラス玉を置く場所をループさせる

            if carry_stone_put == 6: # P1のスコアの穴はスキップする
                continue

            if 0 <= carry_stone_put <= 5: # P1の穴
                Mancala_P1_holes[carry_stone_put] += 1 # P1の穴に石を追加するように修正
            elif carry_stone_put == 13: # P2のスコアの穴
                Mancala_score_P2 += 1
            elif 7 <= carry_stone_put <= 12: # P2の穴 (Mancala_P2_holesのインデックスに変換)
                p2_board_idx = carry_stone_put - 7
                Mancala_P2_holes[p2_board_idx] += 1

            stones_in_hand -= 1 #手のガラス玉の中からガラス玉を減少させる

        # P2のキャプチャルールの実装
        # 最後の石が自分の空の穴に落ち、かつ相手の対応する穴に石がある場合
        if stones_in_hand == 0 and 7 <= carry_stone_put <= 12:
            p2_actual_hole_idx = carry_stone_put - 7
            if Mancala_P2_holes[p2_actual_hole_idx] == 1: # 元々空だった場合（今1つ入っている）
                opposite_hole_index_P1 = 5 - p2_actual_hole_idx # P1側の対応する穴のインデックス
                if Mancala_P1_holes[opposite_hole_index_P1] > 0: # 相手の穴に石がある場合
                    # 石を捕獲し、P2のスコアに入れる
                    Mancala_score_P2 += Mancala_P1_holes[opposite_hole_index_P1] + 1 # 相手の穴の石 + 最後に置いた石
                    Mancala_P1_holes[opposite_hole_index_P1] = 0 # 相手の穴を空にする
                    Mancala_P2_holes[p2_actual_hole_idx] = 0 # 自分が最後に置いた穴を空にする

        if carry_stone_put == 13: # 最後の石がP2の得点ゾーンに入った場合
            extra_turn = True
        return True, extra_turn

def winner(): #勝利判定
    if Mancala_score_P1 > Mancala_score_P2:
        print("P1の勝利！")
    elif Mancala_score_P1 < Mancala_score_P2:
        print("P2の勝利！")
    else:
        print("引き分け！")

def play_game():
    global Mancala_P1_holes, Mancala_P2_holes, Mancala_score_P1, Mancala_score_P2
    global p2_q_agent # Q学習エージェントにアクセスできるようにする

    # ゲームの状態を初期化
    Mancala_P1_holes = [4, 4, 4, 4, 4, 4]
    Mancala_P2_holes = [4, 4, 4, 4, 4, 4]
    Mancala_score_P1 = 0
    Mancala_score_P2 = 0

    current_player = "P1"
    show_board()

    # Q-tableをロードし、AIの探索率を0に設定（最適な行動を選択させるため）
    original_epsilon = p2_q_agent.epsilon # 元のエプシロンを保存
    p2_q_agent.epsilon = 0.0 # 訓練済みAIは探索を行わず、常に最適な行動を選択するようにする
    p2_q_agent.load_q_table() # 訓練済みQテーブルをロード

    while sum(Mancala_P1_holes) > 0 and sum(Mancala_P2_holes) > 0:
        if current_player == "P1":
            print("P1のターンです")
            player_moved = False
            while not player_moved:
                try:
                    player_choice = int(input("P1: 動かしたい穴の番号を入力してください (0-5)。"))
                    moved_successfully, extra_turn = move("P1", player_choice)
                    if moved_successfully:
                        player_moved = True
                        if not extra_turn: # 追加ターンがない場合のみプレイヤーを交代
                            current_player = "P2"
                except ValueError:
                    print("無効な入力です。0から5の整数を入力してください。")
            show_board()
            if extra_turn:
                print("P1は追加ターンを得ました！")

        elif current_player == "P2": # AI (Q学習エージェント) のターン
            print("P2のターンです (AI)") # P2がAIであることを示す
            player_moved = False
            while not player_moved:
                state = p2_q_agent.get_state()
                valid_moves = [i for i, stones in enumerate(Mancala_P2_holes) if stones > 0]

                if not valid_moves:
                    print("P2に有効な手がありません。ゲーム終了です。")
                    break # ゲームを終了

                # Q学習エージェントに行動を選択させる
                player_choice = p2_q_agent.choose_action(state, valid_moves, is_training=False)

                if player_choice == -1:
                    print("P2(AI)の選択が無効でした。ゲーム終了です。")
                    break

                print(f"P2(AI)は穴 {player_choice} を選択しました。")
                moved_successfully, extra_turn = move("P2", player_choice)

                if moved_successfully:
                    player_moved = True
                    if not extra_turn: # 追加ターンがない場合のみプレイヤーを交代
                        current_player = "P1"
                else:
                    print("P2(AI)の選択が無効でした。再度選択します。")
            show_board()
            if extra_turn:
                print("P2は追加ターンを得ました！")

        # どちらかのサイドの穴が空になったらゲーム終了
        if sum(Mancala_P1_holes) == 0 or sum(Mancala_P2_holes) == 0:
            break

    print("ゲーム終了！")
    # 最終的なボードを再表示する前に、残りの石をスコアに移動
    if sum(Mancala_P1_holes) > 0:
        print("P1側の穴に残った石をP1のスコアに追加します。")
        Mancala_score_P1 += sum(Mancala_P1_holes)
        Mancala_P1_holes = [0] * 6
    if sum(Mancala_P2_holes) > 0:
        print("P2側の穴に残った石をP2のスコアに追加します。")
        Mancala_score_P2 += sum(Mancala_P2_holes)
        Mancala_P2_holes = [0] * 6

    show_board() # 最終的なボードとスコアを表示
    winner()

    p2_q_agent.epsilon = original_epsilon # エプシロンを元の値に戻す
```

```python
import collections
import pickle # pickleモジュールを追加

class QLearningAgent:
    def __init__(self, alpha=0.1, gamma=0.9, epsilon=0.1):
        self.q_table = collections.defaultdict(lambda: collections.defaultdict(float)) # Qテーブル: {state: {action: q_value}}
        self.alpha = alpha  # 学習率
        self.gamma = gamma  # 割引率
        self.epsilon = epsilon # 探索率 (epsilon-greedy)

    def get_state(self):
        """現在のゲームボードの状態をタプルとして返す"""
        return (tuple(Mancala_P1_holes), tuple(Mancala_P2_holes), Mancala_score_P1, Mancala_score_P2)

    def choose_action(self, state, valid_moves, is_training=True):
        """ε-greedy法で行動を選択する"""
        if not valid_moves:
            return -1 # 有効な手がない場合

        if is_training and random.uniform(0, 1) < self.epsilon: # 探索 (exploration)
            return random.choice(valid_moves)
        else: # 活用 (exploitation)
            q_values = {move: self.q_table[state][move] for move in valid_moves}
            # 最大Q値を持つ行動を選択
            max_q = max(q_values.values())
            best_moves = [move for move, q in q_values.items() if q == max_q]
            return random.choice(best_moves)

    def learn(self, state, action, reward, next_state):
        """Q値を更新する"""
        old_q = self.q_table[state][action]
        # 次の状態の最大Q値を見つける (有効な手がない場合は0)
        # QLearningAgentのコンテキストからMancala_P2_holesにアクセスできるように変更
        # この実装ではMancala_P2_holesがグローバル変数であることを前提としています。
        # 厳密には、agent.get_state()で取得したnext_stateからP2の穴の状態を抽出するべきですが、
        # 現在の設計に合わせてグローバル変数に直接アクセスします。
        temp_p2_holes = next_state[1] # next_stateからP2の穴の状態を取得
        next_state_valid_moves = [i for i, stones in enumerate(temp_p2_holes) if stones > 0]

        if next_state_valid_moves:
            max_next_q = max([self.q_table[next_state][next_action] for next_action in next_state_valid_moves])
        else:
            max_next_q = 0.0

        # Q学習の更新式
        self.q_table[state][action] = old_q + self.alpha * (reward + self.gamma * max_next_q - old_q)

    def save_q_table(self, filename="q_table.pkl"):
        """Qテーブルをファイルに保存する"""
        with open(filename, 'wb') as f:
            pickle.dump(dict(self.q_table), f)
        print(f"Qテーブルを '{filename}' に保存しました。")

    def load_q_table(self, filename="q_table.pkl"):
        """ファイルからQテーブルをロードする"""
        try:
            with open(filename, 'rb') as f:
                loaded_q_table_dict = pickle.load(f)
            self.q_table = collections.defaultdict(lambda: collections.defaultdict(float), loaded_q_table_dict)
            print(f"Qテーブルを '{filename}' からロードしました。サイズ: {len(self.q_table)}")
            return True
        except FileNotFoundError:
            print(f"'{filename}' が見つかりませんでした。新しいQテーブルを作成します。")
            return False

# Q学習エージェントのインスタンスを作成
p2_q_agent = QLearningAgent(alpha=0.2, gamma=0.9, epsilon=0.3) # パラメータは調整可能
```

```python
def play_game_rl(agent, num_episodes=1000, is_training=True):
    global Mancala_P1_holes, Mancala_P2_holes, Mancala_score_P1, Mancala_score_P2

    # P1のAI (当面はランダム) または人間プレイヤー
    def p1_ai_move():
        valid_moves = [i for i, stones in enumerate(Mancala_P1_holes) if stones > 0]
        if valid_moves:
            return random.choice(valid_moves)
        return -1

    for episode in range(num_episodes):
        # ゲームの状態を初期化
        Mancala_P1_holes = [4, 4, 4, 4, 4, 4]
        Mancala_P2_holes = [4, 4, 4, 4, 4, 4]
        Mancala_score_P1 = 0
        Mancala_score_P2 = 0

        current_player = "P1"

        if not is_training and episode == 0: # トレーニング時以外は初回のみボード表示
            print("--- 新しいゲーム開始 ---")
            show_board()

        while sum(Mancala_P1_holes) > 0 and sum(Mancala_P2_holes) > 0:
            if current_player == "P1":
                # P1は人間がプレイ、またはP1_AIがプレイ
                if is_training:
                    # トレーニング中はP1もAI (ランダム)
                    player_choice = p1_ai_move()
                    if player_choice == -1:
                        break # P1に有効な手がなければゲーム終了
                    moved_successfully, extra_turn = move("P1", player_choice)
                else:
                    # P1が人間の場合は、入力プロンプトを表示
                    player_moved = False
                    while not player_moved:
                        try:
                            player_choice = int(input("P1: 動かしたい穴の番号を入力してください (0-5)。"))
                            moved_successfully, extra_turn = move("P1", player_choice)
                            if moved_successfully:
                                player_moved = True
                        except ValueError:
                            print("無効な入力です。0から5の整数を入力してください。")

                if moved_successfully:
                    if not is_training: # トレーニング時以外はボードを表示
                        show_board()
                        if extra_turn:
                            print("P1は追加ターンを得ました！")
                    if not extra_turn:
                        current_player = "P2"


            elif current_player == "P2": # AI (Q学習エージェント) のターン
                state = agent.get_state()
                valid_moves = [i for i, stones in enumerate(Mancala_P2_holes) if stones > 0]

                if not valid_moves:
                    break # P2に有効な手がなければゲーム終了

                action = agent.choose_action(state, valid_moves, is_training)

                # 行動実行前のP2のスコアを記録
                p2_score_before_move = Mancala_score_P2

                moved_successfully, extra_turn = move("P2", action)

                if moved_successfully:
                    next_state = agent.get_state()

                    # 報酬の計算: スコアの増加分を主な報酬とする
                    reward = (Mancala_score_P2 - p2_score_before_move) * 10 # スコア獲得を重視
                    if extra_turn: # 追加ターン獲得にボーナス
                        reward += 5
                    # 相手の石を奪うことにもボーナスを追加することも検討可能

                    if is_training:
                        agent.learn(state, action, reward, next_state)
                    else: # トレーニング時以外はAIの選択とボードを表示
                        print(f"P2(AI)は穴 {action} を選択しました。")
                        show_board()
                        if extra_turn:
                            print("P2は追加ターンを得ました！")

                    if not extra_turn:
                        current_player = "P1"


            # どちらかのサイドの穴が空になったらゲーム終了
            if sum(Mancala_P1_holes) == 0 or sum(Mancala_P2_holes) == 0:
                break

        # ゲーム終了時の処理（残った石をスコアに加算）
        if sum(Mancala_P1_holes) > 0:
            if not is_training:
                print("P1側の穴に残った石をP1のスコアに追加します。")
            Mancala_score_P1 += sum(Mancala_P1_holes)
            Mancala_P1_holes = [0] * 6
        if sum(Mancala_P2_holes) > 0:
            if not is_training:
                print("P2側の穴に残った石をP2のスコアに追加します。")
            Mancala_score_P2 += sum(Mancala_P2_holes)
            Mancala_P2_holes = [0] * 6

        # 最終報酬: ゲームの勝敗をQテーブルの更新に利用
        if is_training:
            final_reward_p2 = Mancala_score_P2 - Mancala_score_P1 # P2のスコア - P1のスコア
            # ゲーム終了時のQ値更新 (Terminal state) - 最後の行動に対する報酬を学習させる
            # 実際には、エピソードの最後に学習される最後のステップのQ値がこの最終報酬を反映するように設計される
            # ただし、ここでは単純化のため、最終報酬はゲーム中の報酬計算に含めるか、別途処理が必要
            # Q学習は通常、ステップごとの報酬を次状態のQ値に割引して含めるため、ここでは明示的な「最終報酬」の処理は省略
            pass

        if is_training and (episode + 1) % 100 == 0:
            print(f"Episode {episode + 1}/{num_episodes} 完了。 Q-tableサイズ: {len(agent.q_table)}")

        if not is_training:
            print("--- ゲーム終了！ ---")
            show_board()
            winner()

    if is_training:
        print("--- 強化学習トレーニング完了！ ---\n")
        print(f"最終的なQ-tableのサイズ: {len(agent.q_table)}")
        agent.save_q_table() # トレーニング完了後にQテーブルを保存
```

```python
# Q学習エージェントをトレーニングする
print("Q学習エージェントのトレーニングを開始します...")
play_game_rl(p2_q_agent, num_episodes=200000, is_training=True)

# トレーニング後のAIと対戦する (P1は人間プレイヤー)
print("\n--- トレーニング済みAIとの対戦を開始します (P1: 人間, P2: AI) ---")
play_game_rl(p2_q_agent, num_episodes=1, is_training=False)
```
